```python
penguin_data['island'].value_counts().plot(kind='bar',color=['#d5e0fe','#656371','#ff7369'])
```

```python
sns.set_style('white')
penguin_data['species'].value_counts().plot(kind='barh',color=['#6baddf','#01193f','#d2b486'])
```

```python
sns.swarmplot(x=penguin_data.island,y=penguin_data.culmen_length_mm,hue=penguin_data.species)
```

```python
sns.set_style('dark')
sns.boxplot(x=penguin_data.species,y=penguin_data.body_mass_g,hue=penguin_data.sex)
```

```python
sns.set_style('dark')
sns.scatterplot(x=penguin_data.culmen_length_mm,y=penguin_data.culmen_depth_mm,hue=penguin_data.species)
```

```python
correlation_matrix=penguin_data.corr()
correlation_matrix
```

```python
sns.set_style('dark')
sns.heatmap(correlation_matrix,annot=True,linecolor='white',linewidths=5,cmap="YlGnBu")
```

```python
sns.set_style(style='white')
sns.pairplot(data=penguin_data,hue='species',palette=['#6baddf','#01193f','#d2b486'])
```
